### **1. Trasformare cicli in list comprehension**

Fai tre esempi di cicli producono una lista e poi riscrivi lo stesso codice usando la list comprehension

```python
# es1: aumentare gli elementi di una lista di 1
numeri = [1, 4, 3, 8, 5,64]
lettere = "adgdcaaba"
quadrati = []

for n in numeri:
    quadrati.append(n + 1)

# con lista con list comprehension

quadrati = [n + 1 for n in numeri]

# es2: trovare tutti in numeri pari
pari = []

for n in numeri:
    if n % 2 == 0:
        pari.append(n)

# con lista con list comprehension

pari = [n for n in numeri if n % 2 == 0]

# es3:
lettera = []

for n in lettere:
    if n == "a":
        lettera.append(n)

# con lista con list comprehension

lettera = [n for n in lettere if n == "a"]


```

1. **Crea un vettore** contenente tutti i numeri primi compresi tra **0 e 10**  
   _(Puoi scriverli direttamente nell'array)._

2. **Conta quanti numeri ci sono nel vettore** utilizzando la funzione `len()`.  **Ottieni lo stesso numero** accedendo all'attributo `.size` del vettore.

3. Quale pensi sia il tipo di dato (`dtype`) del vettore?  Prova a rispondere senza eseguire il codice e Verifica la tua risposta accedendo all'attributo `.dtype` del vettore

4. Scrivi l'array usando una list comprehension che controlla che i numeri siano primi. 

```python
import numpy as np

# 
primi = np.array([2, 3, 5, 7])

# 2
contlen = len(primi)
contsize = primi.size

print("Vettore primi:", primi)
print("Lunghezza len():", contlen)
print("Lunghezza .size:", contsize)

# 3
# siccome tutti i numeri sono interi senza virgola, il dtype sarà int
print("Tipo dato:", primi.dtype)

# 4. List comprehension con comandi semplici
vettore = np.array([n for n in range(11) if n>1 and all(n % i != 0 for i in range(2,n))])

print("Vettore della list comprehension:", vettore)
```
