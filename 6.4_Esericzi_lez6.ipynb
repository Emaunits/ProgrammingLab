# 5. Analisi di un dataset di video giochi
1. Quanti videogiochi sono stati pubblicati
2. Quali sono i generi più popolari? Mostralo con un bar plot
3. Plotta l'evoluzione del numero di giochi pubblicati nel tempo
4. Prova a riprodurre questo plot 




```python
# Import dei pacchetti
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Caricamento del dataset
url = 'https://zenodo.org/record/5898311/files/vgsales.csv'
df = pd.read_csv(url)
df.head()
```

```python
print(f"Numero pubblicazioni: {len(df)}")

print(f"videogiochi unici: {df['Name'].nunique()}")
```

```python
generi_popolari = df['Genre'].value_counts()
print(generi_popolari)

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Genre', order=df['Genre'].value_counts().index, palette='viridis')
plt.xlabel('Genere')
plt.ylabel('giochi pubblicati')
plt.show()
```

```python
x = df.groupby('Year').size()

plt.figure(figsize=(12, 6))
x.plot(kind='line', marker='o', color='b')
plt.xlabel('Anno')
plt.ylabel('Numero di giochi')
plt.grid(True)
plt.show()
```

```python
vendite = df.groupby('Year')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()


plt.figure(figsize=(14, 7))
sns.lineplot(data=vendite)
plt.xlabel('Anno')
plt.ylabel('Vendite')
plt.grid(True)
plt.legend(['Nord America', 'Europa', 'Giappone', 'Altri paesi'])
plt.show()
```
